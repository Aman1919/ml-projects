```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("./data/data.csv")
df.info()
df.describe()
```

```python
sns.heatmap(df.notnull())
```

```python
df.drop(["Unnamed: 32","id"],axis=1, inplace=True)
df.head()
```

```python
df.diagnosis = [ 1 if value == "M" else 0 for value in df.diagnosis]
df.head()
```

```python
df.diagnosis = df['diagnosis'].astype("category",copy=False)
df.diagnosis.value_counts().plot(kind = "bar")
```

```python
# divide into target variable and predictors
y = df['diagnosis']
x = df.drop(['diagnosis'],axis=1)
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

# normalise the features data
# create a scaler object
scaler  = StandardScaler()

X_scaled = scaler.fit_transform(x)

X_scaled
```

```python
from sklearn.model_selection import train_test_split

## split the test and train data
train_x,test_x,train_y,test_y  = train_test_split(X_scaled ,y , test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

##train the model

lr = LogisticRegression()

#train the model on the training data
lr.fit(train_x,train_y)

y_predict = lr.predict(test_x)
y_predict
```

```python
#evaluation of the model
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(test_y,y_predict)
accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,y_predict))
```
